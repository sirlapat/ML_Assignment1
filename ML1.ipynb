```python
df = pd.read_csv('/content/drive/MyDrive/D3.csv')
X1 = df.values[:,0]
X2 = df.values[:,1]
X3 = df.values[:,2]
Y = df.values[:,3]

# Plot X1
plt.scatter(X1, Y, color='indigo',marker='*')
plt.grid(True)
plt.rcParams["figure.figsize"] = (6, 6)
plt.xlabel('plot of variable X1')
plt.ylabel('plot of y')
plt.title('model graph')

plt.show()

# Plot X2
plt.scatter(X2, Y, color='indigo',marker='*')
plt.grid(True)
plt.rcParams["figure.figsize"] = (6, 6)
plt.xlabel('plot of variable X2')
plt.ylabel('plot of y')
plt.title('model graph')
plt.show()

# Plot X3
plt.scatter(X1, Y, color='indigo',marker='*')
plt.grid(True)
plt.rcParams["figure.figsize"] = (6, 6)
plt.xlabel('plot of variable X3')
plt.ylabel('plot of y')
plt.title('model graph')

plt.show()
```

```python
m = len(X1) #dim of data column
X0 = np.ones((m,1)) #creates X0 filled with 1s
X1 = X1.reshape(m,1) #transposes explanatory columns
X2 = X2.reshape(m,1)
X3 = X3.reshape(m,1)
Y = Y.reshape(m,1)
```

```python
def compute_cost(X, y, theta):

    predictions = X.dot(theta)
    errors = np.subtract(predictions, y)
    sqrErrors = np.square(errors)
    J = 1 / (2 * m) * np.sum(sqrErrors)
    return J

def gradient_descent(X, y, theta, alpha, iterations):

    m = len(y)  # Number of training examples
    cost_history = np.zeros(iterations)
    for i in range(iterations):
        predictions = X.dot(theta)
        errors = np.subtract(predictions, y)
        sum_delta = (alpha / m) * X.transpose().dot(errors)
        theta -= sum_delta
        cost_history[i] = compute_cost(X, y, theta)

    return theta, cost_history
```

```python

iterations = 3300
alpha = 0.01
X = np.column_stack((X0, X1))
theta = np.zeros((2, 1))
theta, loss_history = gradient_descent(X, Y, theta, alpha, iterations)
print('Theta for X1:', theta[1])
print('Theta for X0 (bias):', theta[0])
```

```python
plt.scatter(X1, Y, label='Data Points')
plt.plot(X1, X @ theta, color='indigo', label='Linear Regression')
plt.xlabel('X1 path')
plt.ylabel('Y path')
plt.title('Regression of X1')
plt.legend()
```

```python
plt.plot(range(iterations), loss_history, label='Loss')
plt.grid()
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.title('X1 Loss')
plt.legend()
```

```python
iterations = 3300
alpha = 0.08
X = np.column_stack((X0, X2))
theta = np.zeros((2, 1))
theta, loss_history = gradient_descent(X, Y, theta, alpha, iterations)
print('Theta for X2:', theta[1])
print('Theta for X0 (bias):', theta[0])
```

```python
plt.scatter(X2, Y, label='Data')
plt.plot(X2, X.dot(theta), color='indigo', label='Linear Regression')
plt.xlabel('X2 path')
plt.ylabel('Y path)')
plt.title('Regression of X2')
plt.legend()
```

```python
plt.plot(range(1, iterations + 1), loss_history, label='Loss History')
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.title('X2 Loss')
plt.grid(True)
plt.legend()
```

```python
iterations = 3300
alpha = 0.01
X = np.hstack((X0,X3))
theta = np.zeros((2,1))
theta, loss_history = gradient_descent(X, Y, theta, alpha, iterations)
print('Theta for X3:', theta[1])
print('Theta for X0 (bias):', theta[0])
```

```python
plt.scatter(X3, Y)
plt.plot(X3, X.dot(theta), color = 'indigo', label = 'Linear Regression Fit')
plt.xlabel('X3 path')
plt.ylabel('Y path')
plt.title('Regression of X3')
plt.legend()
plt.show()
```

```python
plt.plot(range(1, iterations + 1), loss_history, color='orangered')
plt.grid()
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.title('Loss X3')
plt.show()
```

```python
iterations = 3300
alpha = 0.05
X = np.hstack((X0,X1,X2,X3))
theta = np.ones((4,1)) * 0.5

theta, loss_history = gradient_descent(X, Y, theta, alpha, iterations)
print('Final theta for X_ALL:', theta)
print('Final loss:', loss_history[-1])
```

```python
plt.plot(range(1, iterations + 1), loss_history, color='red', marker='+', markersize=2)
plt.grid(linestyle='--')
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.title('X_all Loss, alpha = 0.05')
plt.tight_layout()
X_test = np.array([[1,1,1,1], [1,2,0,4], [1,3,2,1]])
predictions = X_test.dot(theta)

for i, prediction in enumerate(predictions):
  print(f'Prediction {i+1} ({X_test[i,1:]}) = {prediction}')
```
